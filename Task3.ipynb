```python
f = open("pr1.txt", "r")
df=f.read()
f2 = open("pr2.txt", "r")
df2=f2.read()
```

```python
import nltk
nltk.download('averaged_perceptron_tagger')
```

```python
import re
import unicodedata
import nltk
from nltk.corpus import stopwords
from collections import Counter
```

```python
def basic_clean(text):
    

    wnl = nltk.stem.WordNetLemmatizer()
    stopwords = nltk.corpus.stopwords.words('english') 
    words = re.sub(r'[^\w\s]', '', text).split()
    return [wnl.lemmatize(word) for word in words if word not in stopwords]
```

```python
words=basic_clean(df)
words2=basic_clean(df2)
```

```python
unigram=nltk.ngrams(words,1)
unigramFrequency=Counter(unigram)
valuesOfUnigrams=list(unigramFrequency.values())
unigramlist=list(unigramFrequency)

for x in range(0,len(unigramlist)):
    if(valuesOfUnigrams[x]>0):
        print(unigramlist[x],"is used",valuesOfUnigrams[x],"times")
count1=len(unigramlist)
print("1. corpusta",count1,"adet unigram bulunuyor")
```
